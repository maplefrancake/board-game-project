```python
knn_df_scaled = StandardScaler().fit_transform(knn_game_df[[
                           'GameWeight', 'MfgAgeRec','Cat:Thematic',
                           'Cat:Strategy','Cat:War','Cat:Family','Cat:CGS','Cat:Abstract',
                           'Cat:Party', 'Cat:Childrens']])


# knn_df_scaled = StandardScaler().fit_transform(knn_game_df[["YearPublished",'NumUserRatings','AvgRating','MinPlayers',
#                            'GameWeight', 'MfgAgeRec','MfgPlaytime','Cat:Thematic',
#                            'Cat:Strategy','Cat:War','Cat:Family','Cat:CGS','Cat:Abstract',
#                            'Cat:Party', 'Cat:Childrens']])
knn_df_scaled
```

```python
knn_df_transformed = pd.DataFrame(knn_df_scaled, columns=[
                           'GameWeight', 'MfgAgeRec','Cat:Thematic',
                           'Cat:Strategy','Cat:War','Cat:Family','Cat:CGS','Cat:Abstract',
                           'Cat:Party', 'Cat:Childrens'])
knn_df_transformed
```

```python
from sklearn.neighbors import NearestNeighbors
neigh = NearestNeighbors(n_neighbors=3)
neigh.fit(knn_df_transformed.iloc[:,1:10])
```

```python
knn_game_df[knn_game_df["Name"].str.contains("Chess")]
```

```python
game = knn_df_transformed.iloc[150][1:10]
game.to_numpy()
```

```python
closest_neighbor = neigh.kneighbors(game.to_numpy().reshape(1, -1))
print(closest_neighbor)
print(f"Nearest neighbor: ")
# print(f"Distance to nearest neighbor: {closest_neighbor[0]}")
print(knn_game_df.iloc[closest_neighbor[1][0][1],1])
print(knn_game_df.iloc[closest_neighbor[1][0][2],1])

s = knn_game_df.iloc[closest_neighbor[1][0][1],:3]
df = pd.DataFrame(s)
df.to_html()
```
